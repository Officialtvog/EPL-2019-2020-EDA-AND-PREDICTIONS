```python
team_results = []
for team in teams:
    # Get the data for that team
    teams_data = df11[df11['teamId'] == team]
    
    wins = teams_data['wins'].sum()
    draws = teams_data['draws'].sum()
    losses = teams_data['loses'].sum()
    scored = teams_data['scored'].sum()
    conceded = teams_data['missed'].sum()
    games = wins + draws + losses
    points = (3 * wins) + draws
    goal_difference = scored - conceded
    
    team_results.append([team, games, wins, draws, losses, scored, conceded, goal_difference, points])

epl_table = pd.DataFrame(team_results, columns=['Team', 'P', 'W', 'D', 'L', 'F', 'A', 'GD', 'Points'])
epl_table.sort_values(by=['Points', 'GD', 'F'], ascending=False, inplace=True, ignore_index=True)
epl_table.set_index(POSITIONS, inplace=True)
epl_table.head(20)
```

```python

df11['xGround'] = df11['xG'].apply(lambda x: round(x))
df11['xGAround'] = df11['xGA'].apply(lambda x: round(x))
df11['xwin'] = df11.apply(lambda x: 1 if x['xGround'] > x['xGAround'] else 0, axis=1)
df11['xdraw'] = df11.apply(lambda x: 1 if x['xGround'] == x['xGAround'] else 0, axis=1)
df11['xloss'] = df11.apply(lambda x: 1 if x['xGround'] < x['xGAround'] else 0, axis=1)
```

```python
# Get the results for each team
new_epl_results = []
for team in teams:
    # Get the data for that team
    teams_data = df11[df11['teamId'] == team]
    
    wins = teams_data['xwin'].sum()
    
    draws = teams_data['xdraw'].sum()
    
    scored = teams_data['xGround'].sum()
    
    losses = teams_data['xloss'].sum()
    
    conceded = teams_data['xGAround'].sum()
    
    games = wins + draws + losses
    
    points = (3 * wins) + draws
    
    goal_difference = scored - conceded
    
    new_epl_results.append([team, games, wins, draws, losses, scored, conceded, goal_difference, points])

new_epl_table = pd.DataFrame(new_epl_results, columns=['Team', 'P', 'W', 'D', 'L', 'F', 'A', 'GD', 'Points'])

new_epl_table.sort_values(by=['Points', 'GD', 'F'], ascending=False, inplace=True, ignore_index=True)

new_epl_table.set_index(POSITIONS, inplace=True)

new_epl_table.head(20)
```
